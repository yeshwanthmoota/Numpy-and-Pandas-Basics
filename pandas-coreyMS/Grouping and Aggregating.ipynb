### Note:

A median(statistics) on the data gives more accurate information as compared to a mean(statistics) because a few outliers in the data can have a huge impact on the mean of the data where as it will have no impact on the median of the data.

How is median calculated:
Here's an example to illustrate the calculation of the median:

Let's say we have the following set of data: 5, 2, 7, 1, 1000, 3, 6, 4.

1. Arrange the data in ascending order: 1, 2, 3, 4, 5, 6, 7, 1000.
2. The number of data points is 8 (n = 8). 
3. Since n is even, we need to find the average of the two middle values.
    1. The two middle values are at indices (n / 2) = 4 and (n / 2) + 1 = 5.
    2. The values at these indices are 4 and 5.
    3. Calculate the average: (4 + 5) / 2 = 4.5.
So, the median of the given data set is 4.5.

If the number of data points is odd, you simply need to find the middle value, i.e. (n+1)/2 'th value directly without averaging.


And the mean of the data 5, 2, 7, 1, 1000, 3, 6, 4 is (5 + 2 + 7 + 1 + 1000 + 3 + 6 + 4)/8 = 128.5
which gives no real information about our data because it had an outlier i.e. '1000' which affected the whole mean.


That is the reason median gives more accurate information as compared to mean

### Note:

When you use filtered_df = df[filter1 and filter2] instead of filtered_df = df[filter1 & filter2], you will encounter a ValueError with the message "The truth value of a Series is ambiguous."

The reason for this error is that the and operator (and) is not suitable for combining multiple filter conditions in pandas. The and operator expects single boolean values on both sides, but in this case, filter1 and filter2 are Series objects (resulting from the comparison operations) rather than single boolean values.

To combine multiple filter conditions in pandas, you need to use the & operator instead. The & operator performs an element-wise logical "and" operation between two Series, producing a new Series of boolean values indicating the filtered rows.

Therefore, it is important to use & instead of and when combining filters in pandas.

```python
filt1 = (df["LanguageWorkedWith"].str.contains("Python", case=False, na=False))
filt2 = (df["Country"] == "India")
df.loc[filt1&filt2][["Country", "LanguageWorkedWith"]]
```

```python
python_df.rename(columns={'Country': 'NumRespondents', 'LanguageWorkedWith': 'NumKnowsPython'}, inplace=True)
python_df
```

```python
python_df['PctKnowsPython'] = (python_df['NumKnowsPython']/python_df['NumRespondents']) * 100
python_df
```

```python
python_df.sort_values(by='PctKnowsPython', ascending=False, inplace=True)
python_df.head(50)
```
