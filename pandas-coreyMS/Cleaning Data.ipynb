```python
import numpy as np
import pandas as pd
```

```python
people = {
    'first': ['Corey', 'Jane', 'John', 'Chris', np.nan, None, 'NA'], 
    'last': ['Schafer', 'Doe', 'Doe', 'Schafer', np.nan, np.nan, 'Missing'], 
    'email': ['CoreyMSchafer@gmail.com', 'JaneDoe@email.com', 'JohnDoe@email.com', None, np.nan, 'Anonymous@email.com', 'NA'],
    'age': ['33', '55', '63', '36', None, None, 'Missing']
}
```

```python
df = pd.DataFrame(people)
df
```

```python
df.dropna() # used to drop all the rows with 'NaN' or 'np.nan' or 'None' values
```

```python
# default arguments for dropna() are axis='index', how='any'
df.dropna(axis='index', how='any')
```

-> For axis we have 2 options i.e. 'index' (drops rows that have NaN values) and 'columns' (drops columns that have NaN values)  
-> For how we have options any 'any' if any of the values in the axis are null values then drop the row or column wrt axis, 'all' where all of the values in the axis are null values then drop the row or column wrt axis


```python
df.dropna(axis='index', how='all')
```

We also have another argument i.e. subset=[] where we can write column names we absolutely don't want to null while the axis is 'index'   
i.e. for a row to be dropped if how='all' any of the subset values should have non-NaN values  
if how='any' then if any of the column names passed in the subset have a NaN value then the row is dropped

```python
df.dropna(axis='index', how='all', subset=['email']) # if email is non-NaN it will not drop the row
```

```python
df.dropna(axis='index', how='all', subset=['last', 'email']) # if any of 'last' or 'email' is there it will not drop the row
```

```python
df.dropna(axis='index', how='any', subset=['last', 'email']) # if any of 'last' or 'email' is NaN it will drop the row
```

We can use the inplace=True to actually change the dataframe.

Now to remove the customized missing values like 'NA' and 'missing'

```python
df.replace('NA', np.nan, inplace=True)
df.replace('Missing', np.nan, inplace=True)
df
```

```python
# We can get a mask of all the NA values in the dataframe using df.isna()
df.isna()
```
